```python
df2 = pd.DataFrame({'c1': [1, 2, 3, 4, 5],
                    'c2': [6, 7, 8, 9, 10]})

df2
```

```python
# 7, 9를 np.nan으로 변경
df2.loc[[1, 3], 'c2'] = np.nan
df2
```

```python
# c2new 열을 추가(c2열 값이 null이 아니면 c2 그렇지 않으면, c1 열 값으로 채움)
df2['c2new'] = np.where(df2['c2'].notnull(), df2['c2'], df2['c1'])
df2
```

```python
for i in df2.index:
    if np.isnan(df2.loc[i, 'c2']):
        df2.loc[i, 'c2new'] = df2.loc[i, 'c1']
    else:
        df2.loc[i, 'c2new'] = df2.loc[i, 'c2']
        
df2
```

### 중복값 여부 확인
`duplicated()` : return boolean series denoting duplicate rows  
`drop_duplicates()` : return data frame with duplicate rows removed

```python
dic = {
    'key1': ['a', 'b', 'b', 'c', 'c'],
    'key2': ['v', 'w','w', 'x', 'y'],
    'col': [1, 2, 3, 4, 5]
}

df = pd.DataFrame(dic)
df
```

`중복데이터 여부 확인 duplicated()`

```python
df.duplicated('key1')
df.duplicated(['key1'])
```

```python
df[df.duplicated(['key1', 'key2'])]
```

`duplicated()`  
`default: keep='first'`

```python
print(df.duplicated(['key1', 'key2'], keep='first'), end="\n\n")
print(df.duplicated(['key1', 'key2'], keep='last'), end="\n\n")
print(df.duplicated(['key1', 'key2'], keep=False))
```

```python
df.drop_duplicates(['key1'], keep='first')
```

```python
df.drop_duplicates(['key1'], keep='last')
```

```python
df.drop_duplicates(['key1'], keep=False)
```
