# xml스크래핑

```python
import urllib.parse as up
```

```python
#매개편수 값에 따른 웹페이지 추출
#http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109 (?부터가 매개변수)
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
myarg = {'stnId':'184'} #딕셔너리 형태로 매개변수 지정
```

```python
params = up.urlencode(myarg) #매개변수를 주소 형태로 결합해주는 인코딩 코드
```

```python
addr = url+"?"+params #최종 접속할 주소로 결합
print(addr)
```

```python
print(urllib.request.urlopen(addr).read()) #주소에서 제공하는 문서 읽어오기(xml형태)
                                           #\x는 문서가 16진수로 저장되어있다는 뜻
res = urllib.request.urlopen(addr).read()
```

```python
text = res.decode("utf-8") #16진수로 저장된 문서를 읽을 수 있는 형태로 변형
print(text)
```

```python
from bs4 import BeautifulSoup #beautifulsoup 모듈 import
# Beautiful Soup is a Python library for pulling data out of HTML and XML files.
# bs4 문서 : https://www.crummy.com/software/BeautifulSoup/bs4/doc/
```

```python
#가상의 html문서를 생성(실습용)
html = """
<html><body>
    <h1 id="title">스크래핑?</h1> 
    <p id="body">웹 페이지 분석</p>
    <p>원하는 부분을 추출</p>
</body></html>
"""
#<h1 id="title">스크래핑?</h1>에서 h1은 태그, id는 속성
```

```python
soup = BeautifulSoup(html,"html.parser") 
#첫번째 변수 : 분석할 문서
#두번째 변수 : parser(파서)->구문분석(parsing)
#parser 종류 : html.parser, lxml(더 가벼움)
```

```python
print(soup.find(id="title").string)
print(soup.find(id="body").string)
```

```python
html = """
<html><body>
<ul>
<li><a href="http://www.naver.com">naver</a></li>
<li><a href="http://www.daum.net">daum</a></li>
</ul>
</body></html>
"""
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
links = soup.find_all("a") #리스트 형태
```

```python
#속성 값을 출력하고 싶을 때    
for i in links:
    print(i.attrs['href'])

#tag에 걸린 text를 출력하고 싶을 때
for i in links:
    print(i.string)
```

```python
html = """
<ul id="lang">
    <li id="ja">java</li>
    <li id="py">python</li>
    <li id="cpp">c++</li>
    <li id="sp">spark</li>
</ul>
"""

soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup.select("li")[1].string)

print(soup.select_one("#py").string)
print(soup.select_one("li#py").string)

print(soup.select_one("ul > #py").string)
print(soup.select_one("ul#lang li#py").string)
print(soup.select_one("li[id='py']").string)

print(soup.select_one("li:nth-of-type(2)").string)
print(soup.select_one("li:nth-child(2)").string)
```
